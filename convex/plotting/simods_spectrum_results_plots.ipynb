```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def plot_spectra(results_dir, datasets, dataset_names):
    plt.figure()

    for dataset, dataset_name in zip(datasets, dataset_names):
        S = np.load(os.path.join(results_dir, f'{dataset}_top_sing_val.npy'))
        plt.semilogy(S, label = dataset_name)

    plt.xlabel('$k$', fontsize = 14)
    plt.ylabel(r'$\sigma_k$', fontsize = 14)
    plt.legend(loc = 'upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize = 14)
    plt.gca().tick_params(axis='both', which='major', labelsize=14)
```

```python
# Render plots in LaTeX
plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')
```

```python
save_dir = './spectra'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

results_dir = os.path.abspath('../simods_spectra_results/')
```
